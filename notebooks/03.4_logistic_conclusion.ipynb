## Comparison Between Baseline and Engineered Logistic Regression Models

To improve the statistical validity of the logistic regression model, diagnostic checks were conducted on the baseline specification and compared with a refined model obtained through feature engineering.

### Baseline Model

The initial model included the predictors:

- `age`
- `ejection_fraction`
- `time`
- `serum_creatinine`

#### Multicollinearity

Variance Inflation Factor (VIF) diagnostics revealed potential multicollinearity:

| Feature | VIF |
|-------|------|
| age | 10.93 |
| ejection_fraction | 9.47 |
| time | 3.35 |
| serum_creatinine | 2.88 |

The VIF values for **age** and **ejection_fraction** were close to or above common thresholds (5–10), indicating possible instability in coefficient estimates.

#### Linearity Diagnostics

Plots of predictors against the logit of predicted probabilities suggested:

- A moderate linear relationship for **time**.
- Slight curvature for **age**.
- Possible non-linear patterns for **serum_creatinine**.
- A somewhat non-linear decreasing trend for **ejection_fraction**.

These observations motivated further feature engineering.

---

### Engineered Model

The refined model applied the following transformations:

- Centering of continuous variables:
  - `age_centered`
  - `ejection_fraction_centered`
- Construction of an interaction term:
  - `sodium_creatinine_interaction`
- Retention of the predictor:
  - `time`

#### Multicollinearity

After transformation, the VIF diagnostics improved substantially:

| Feature | VIF |
|-------|------|
| time | 3.35 |
| sodium_creatinine_interaction | 3.34 |
| age_centered | 1.02 |
| ejection_fraction_centered | 1.01 |

All predictors now exhibit **very low multicollinearity**, with VIF values close to 1.

#### Linearity Diagnostics

Visual inspection of logit plots suggests:

- A clearer linear trend for **age_centered**.
- A stable monotonic relationship for **time**.
- Improved structure in the relationship involving **ejection_fraction**.
- The interaction between **sodium** and **creatinine** captures nonlinear risk patterns observed in the baseline specification.

---

### Clinical Relevance of the Interaction

The interaction between **serum sodium and creatinine** is clinically meaningful, as both variables reflect key aspects of cardiovascular and renal function. In heart failure patients, the coexistence of **hyponatremia (low sodium)** and **impaired renal function (high creatinine)** is often associated with worse prognosis, so modeling their interaction helps capture this compounded risk.

---

### Summary

The engineered model improves the baseline specification in several ways:

- **Reduced multicollinearity**, resulting in more stable coefficient estimates.
- **Improved adherence to the linearity assumption in the logit.**
- **More interpretable predictors** through centering and interaction modeling.

Overall, the feature engineering process leads to a more statistically robust logistic regression model while preserving interpretability.
