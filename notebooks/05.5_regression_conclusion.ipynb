# Conclusion

This section summarizes the exploratory and modeling steps performed in this didactic regression exercise using the heart failure clinical dataset. The objective was not only to obtain accurate predictions but also to illustrate the role of feature engineering, statistical inference, and model diagnostics in a typical regression workflow.

---

# Target Selection

Although the dataset was originally designed for classification tasks (predicting `DEATH_EVENT`), we explored whether any continuous variable could serve as a meaningful regression target.

To do so, we evaluated the predictability of each numeric variable using cross-validated R² scores.

The results showed that most original clinical variables exhibit very low predictive power, confirming that the dataset was not originally structured for regression tasks.

However, among the original continuous variables, **ejection fraction** displayed the highest predictive signal and strong clinical relevance. For this reason, it was selected as the regression target.

---

# Feature Engineering

Exploratory analysis suggested that relationships between individual variables were relatively weak. To introduce additional predictive structure, several transformations and interaction terms were constructed:

- `creatinine_log`
- `ejection_creatinine_interaction`
- `sodium_creatinine_interaction`
- `age_centered`

These engineered features substantially improved the relationships between predictors and the target variable.

The correlation analysis revealed that interaction terms exhibited stronger relationships with the target compared to the raw predictors.

This result highlights an important concept:

> Feature engineering can reveal predictive structure that is not apparent in the raw variables.

# Linear Model Specification

Two linear regression models were evaluated.

## Parsimonious Model

This model includes only the strongest predictors identified during exploratory analysis:

- `ejection_creatinine_interaction`
- `creatinine_log`
- `sodium_creatinine_interaction`

## Stepwise Model

A stepwise model selection procedure based on AIC identified the following specification:

- `age_centered`
- `creatinine_log`
- `ejection_creatinine_interaction`
- `sodium_creatinine_interaction`

# Final Remarks

This regression exercise illustrates a typical workflow in applied statistical modeling:

1. Exploratory analysis  
2. Feature engineering  
3. Model specification  
4. Model selection  
5. Diagnostic evaluation  
6. Predictive validation  

While the linear models already achieve reasonable performance, the presence of nonlinear patterns and heteroscedasticity suggests that more flexible models (such as tree-based methods) could potentially further improve predictive accuracy.

Nevertheless, the linear regression framework remains valuable due to its interpretability and its ability to provide insight into the relationships between clinical variables.
